```python
import pandas as pd
```

```python
df=pd.read_csv('police-locals.csv')
```

```python
df
```

# PREPROSSING

```python
df=df.replace('**',0)
df
```

```python
df["all"] = pd.to_numeric(df["all"])
df["white"] = pd.to_numeric(df["white"])
df["non-white	"] = pd.to_numeric(df["non-white"])
df["black"] = pd.to_numeric(df["black"])
df["hispanic"] = pd.to_numeric(df["hispanic"])
df["asian"] = pd.to_numeric(df["asian"])
df
```

```python
df.groupby('city').sum().sort_values("police_force_size",ascending=False)["police_force_size"].head().plot.bar()
```

## Explanation

### This visualization shows that New York has the greatest police force size (greater than 30000)
### The other 4 are significantly lower but not that much if we compare them to one another. All of them are between 5000 to 15000

```python
df[["white","non-white","black","hispanic","asian"]].mean().plot()
```

## Explanation

### This visualization shows that the average percentage of white, non-white, black, hispanic, asian who live in the city
